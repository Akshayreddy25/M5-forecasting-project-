```python
sales = pd.read_csv("../data/sales_train_evaluation.csv")
calendar = pd.read_csv("../data/calendar.csv")

print("Sales shape:", sales.shape)
print("Calendar shape:", calendar.shape)
```

```python
print(sales.columns[:10].tolist())
```

```python
# Step 1 — separate the identifier columns from the day columns
id_cols = ['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id']
day_cols = [col for col in sales.columns if col.startswith('d_')]

print(f"Identifier columns: {len(id_cols)}")
print(f"Day columns: {len(day_cols)}")
```

```python
# Step 2 — melt wide format into long format
sales_long = sales.melt(
    id_vars=id_cols,
    value_vars=day_cols,
    var_name='d',
    value_name='sales'
)

print(f"Shape before melt: {sales.shape}")
print(f"Shape after melt:  {sales_long.shape}")
sales_long.head()
```

```python
print(sales_long.shape)
```

```python
print(calendar.head())
print(calendar.shape)
```

```python
# Merge calendar into sales_long on the 'd' column
sales_cal = sales_long.merge(calendar, on='d', how='left')

print(sales_cal.shape)
sales_cal.head()
```

```python
sales_full = sales_cal.merge(prices, on=['store_id', 'item_id', 'wm_yr_wk'], how='left')

print(sales_full.shape)
sales_full.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sales = pd.read_csv("../data/sales_train_evaluation.csv")
calendar = pd.read_csv("../data/calendar.csv")

print("Sales shape:", sales.shape)
print("Calendar shape:", calendar.shape)
```

```python
sales.head()
```

```python
print(sales.columns[:10].tolist())
```

```python
# Step 1 — separate the identifier columns from the day columns
id_cols = ['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id']
day_cols = [col for col in sales.columns if col.startswith('d_')]

print(f"Identifier columns: {len(id_cols)}")
print(f"Day columns: {len(day_cols)}")
```

```python
# Step 2 — melt wide format into long format
sales_long = sales.melt(
    id_vars=id_cols,
    value_vars=day_cols,
    var_name='d',
    value_name='sales'
)

print(f"Shape before melt: {sales.shape}")
print(f"Shape after melt:  {sales_long.shape}")
sales_long.head()
```

```python
print(sales_long.shape)
```

```python
print(calendar.head())
print(calendar.shape)
```

```python
# Merge calendar into sales_long on the 'd' column
sales_cal = sales_long.merge(calendar, on='d', how='left')

print(sales_cal.shape)
sales_cal.head()
```

```python
print(sales_cal[['date', 'weekday', 'month', 'event_name_1', 'snap_CA', 'sales']].head(10))
```

```python
prices = pd.read_csv("../data/sell_prices.csv")
print(prices.shape)
print(prices.head())
```

```python
sales_full = sales_cal.merge(prices, on=['store_id', 'item_id', 'wm_yr_wk'], how='left')

print(sales_full.shape)
sales_full.head()
```

```python
sales_full['date'] = pd.to_datetime(sales_full['date'])
print(sales_full['date'].dtype)
```

```python
import matplotlib.pyplot as plt

daily_sales = sales_full.groupby('date')['sales'].sum()

plt.figure(figsize=(14, 5))
plt.plot(daily_sales.index, daily_sales.values, linewidth=0.8, color='steelblue')
plt.title('Total Daily Sales — All Stores & Items')
plt.xlabel('Date')
plt.ylabel('Units Sold')
plt.tight_layout()
plt.show()
```

```python
category_sales = sales_full.groupby(['date', 'cat_id'])['sales'].sum().unstack()

category_sales.plot(figsize=(14, 5), linewidth=0.8)
plt.title('Daily Sales by Category')
plt.xlabel('Date')
plt.ylabel('Units Sold')
plt.legend(title='Category')
plt.tight_layout()
plt.show()
```

```python
store_sales = sales_full.groupby('store_id')['sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 5))
plt.bar(store_sales.index, store_sales.values / 1_000_000, color='steelblue', edgecolor='white')
plt.title('Total Sales by Store (All Time)')
plt.xlabel('Store')
plt.ylabel('Total Units Sold (Millions)')
plt.tight_layout()
plt.show()
```

```python
avg_sales_by_day = sales_full.groupby('weekday')['sales'].mean()

day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
avg_sales_by_day = avg_sales_by_day.reindex(day_order)

plt.figure(figsize=(10, 5))
plt.bar(avg_sales_by_day.index, avg_sales_by_day.values, color='steelblue', edgecolor='white')
plt.title('Average Sales by Day of Week')
plt.xlabel('Day')
plt.ylabel('Avg Units Sold')
plt.tight_layout()
plt.show()
```

```python
day_cols = [col for col in sales.columns if col.startswith('d_')]

zero_pct = (sales[day_cols] == 0).sum(axis=1) / len(day_cols) * 100

plt.figure(figsize=(10, 5))
plt.hist(zero_pct, bins=50, color='steelblue', edgecolor='white')
plt.title('% of Zero Sale Days per Item')
plt.xlabel('% Days with Zero Sales')
plt.ylabel('Number of Items')
plt.tight_layout()
plt.show()

print(f"Average zero % across all items: {zero_pct.mean():.1f}%")
```
